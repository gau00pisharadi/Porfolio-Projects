# 1. Data preperation and cleaning

-Importing dataset<br>
-Getting an overview of the dataset<br>
-Delete reduntant columns<br>
-Removing duplicates<br>
-Removing Nan values<br>
-Renaming columns<br>

```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the dataset
df=pd.read_csv(r'C:\Users\Gautam\Desktop\jnotebook\Projects\Zomato\zomato.csv')
df
```

```python
df.head()
```

```python
df.info()
```

```python
#Dropping unnecessary columns
df.drop('listed_in(city)', axis=1, inplace=True)
df.head()
```

```python
#Dropping unnecessary columns
df.drop('menu_item', axis=1, inplace=True)
df.head()
```

```python
#Dropping unnecessary columns
df.drop(['url','reviews_list','address','phone','dish_liked'],axis=1,inplace=True)
df.head()
```

```python
#Count of Nan values
df.isna().sum()
```

```python
# Dropping Nan values
df.dropna(inplace=True)
```

```python
# Inspecting each individual columns. First being 'name'
df.groupby('name').count()
```

```python
# Replacing the '[Ãx][^A-Za-z]+' pattern with empty space
df['name']=df['name'].str.replace('[Ãx][^A-Za-z]+','', regex=True)
df.groupby('name').count().head()
```

```python
#Inspecting the online_order column
df['online_order'].unique()
```

```python
#Inspecting the book_table column
df['book_table'].unique()
```

```python
#Inspecting the rate column
df['rate'].unique()
```

```python
# Replacing the 'NEW' value with '-'
df['rate']=df['rate'].replace('NEW','-')
df['rate'].unique()
```

```python
# Removing the '/5' from the rate column
df["rate"] = df["rate"].str.replace("/5", "")
df['rate'].unique()
```

```python
# converting the rate column into numneric column
df['rate']=pd.to_numeric(df['rate'], errors='coerce').fillna(0)
df['rate'].unique()
print(df.dtypes)
```

```python
#Inspecting the votes column
df['votes'].unique()
```

```python
#Inspecting the location column
df['location'].unique()
```

```python
#Inspecting the rest_type column
df['rest_type'].unique()
```

```python
#Inspecting the cuisines column
df['cuisines'].unique()
```

```python
#Inspecting the approx_cost(for two people) column
df['approx_cost(for two people)'].unique()
```

```python
# Removing the "," from the the column
df['approx_cost(for two people)']=df['approx_cost(for two people)'].str.replace(",","")
df['approx_cost(for two people)'].unique()
```

```python
# converting the column to numeric column
df['approx_cost(for two people)']=pd.to_numeric(df['approx_cost(for two people)'], errors='coerce').fillna(0)
print(df.dtypes)
```

```python
#Inspecting the listed_in(type) column
df['listed_in(type)'].unique()
```

```python
# Number of duplicate columns
df.duplicated().value_counts()
```

```python
# Dropping duplicate columns
df.drop_duplicates(inplace=True)
df.duplicated().value_counts()
```

```python
# Saving the cleaned the dataframe as a csv file
df.to_csv('zomato_clean_data.csv')
```

# 2. Data visualization

-Using matplotlib and seaborn<br>
-Number of restaurants that accept online orders<br>
-Number of restaurants that accept table reservations<br>
-Best location in terms of rating and votes<br>
-Number of restaurants per each type of restaurant<br>
-The cost of in each restaurant<br>
-Votes and rating vs.type of restaurant<br>

```python
df.head()
```

```python
# Countplot of online_order

sns.countplot(x='online_order',data=df)
plt.show()
```

As we can see, 17000+ restaurants are accepting online orders, while only less than 10000 restaurants do not accept online orders. 

```python
# Countplot of book_table
sns.countplot(x='book_table',data=df)
plt.show()
```

As we can see, only 5000 restaurants have table reservations, whil 20000+ restaurants do not have table reservations. 

```python
# Barplot of top 5 locations wiht highest rating
fig=plt.figure(figsize=(16,10))
ax1=plt.subplot(2,1,1)
df_grouped=df.groupby('location')['rate'].median().to_frame('rating').reset_index().sort_values('rating',ascending=False).head()
sns.barplot(x='location',y='rating',data=df_grouped)
plt.show()

# Barplot of top 5 locations wiht highest votes
fig=plt.figure(figsize=(16,10))
ax2=plt.subplot(2,1,2)
df_grouped2=df.groupby('location')['votes'].median().to_frame('votes').reset_index().sort_values('votes',ascending=False).head()
sns.barplot(x='location',y='votes',data=df_grouped2)
plt.show()
```

Hence, "Lavelle Road" location has the highest rating as well as votes as compared to other location.

```python
# Count of restaurants per type of restaurants
plt.figure(figsize=(10,20))
df_grp=df.groupby('rest_type')['name'].count().reset_index().sort_values(by='name',ascending=False).head(50)
sns.barplot(x='name',y='rest_type',data=df_grp)
```

As we can see that more than 8000 restaurants are of quick bytes restaurants, while more than 6000 restaurants are casual dining restaurants.

```python
# The distribution of costs of restaurants
plt.figure(figsize=(12,8))
sns.histplot(df['approx_cost(for two people)'],kde=True,color='r')
```

As we can observe from the plot, most restaurants have an approximate cost of between ₹1 and ₹1000, while the rest have a cost of more than ₹1000.

```python
# Number of votes and rating vs. the restaurant type
plt.figure(figsize=(18,10))
df_grp=df.groupby('rest_type').agg({'votes':'sum','rate':'mean'}).reset_index().sort_values(by='votes',ascending=False).head(5)
sns.barplot(x='rest_type',y='rate',data=df_grp,hue='votes')
```

Hence, most votes were given for casual dining, but the most rating was for casual dining,cafe.
